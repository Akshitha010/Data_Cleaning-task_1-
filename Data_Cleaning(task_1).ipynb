```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
```

```python
# 1. Identifying and handling missing values
# Numeric columns: fill with median; text columns: fill with 'Unknown'
for col in df.columns:
    if df[col].dtype == 'O':  # Object (string) columns
        df[col] = df[col].fillna('Unknown')
    else:
        df[col] = df[col].fillna(df[col].median())
df
```

```python
# 2. Remove duplicate rows
df = df.drop_duplicates()

# 3. Standardize text values
# 'type', 'country', 'rating', 'listed_in' to title case and strip spaces
for col in ['type', 'country', 'rating', 'listed_in']:
    df[col] = df[col].astype(str).str.title().str.strip()
# 'director', 'cast' to strip spaces (keep as string)
for col in ['director', 'cast']:
    df[col] = df[col].astype(str).str.strip()
```

```python
# 4. Convert 'date_added' to datetime (format: dd-mm-yyyy)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.head()
```

```python
# 5. Rename column headers: lowercase, underscores
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
```

```python
# 6. Check and fix data types
# 'release_year' as integer
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')
df.head()
```

```python
# Extract duration as integer (for movies)
df['duration_int'] = df['duration'].str.extract('(\d+)').astype('Int64')
df
```

```python
# 7. Save the cleaned dataset
df.to_csv('cleaned_data.csv', index=False)
print("Cleaned data saved as 'cleaned_data.csv'")
```
